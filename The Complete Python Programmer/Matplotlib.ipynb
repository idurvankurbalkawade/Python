## Importing the matplotlib library

```python
import matplotlib.pyplot as plt
#from matplotlib import pyplot as plt
```

## Line Graph

```python
num=[1,2,3,4,5]
squares=[1,4,9,16,25]
plt.plot(num,squares)
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.title("Numbers and Squares")
plt.savefig("squares.jpg")
plt.show()
```

## Bar Graphs

```python
plt.bar(num,squares,color="red")
plt.savefig("bar.jpg")
plt.show()
```

## Horizontal Bar Graph

```python
plt.barh(num,squares)
plt.savefig("barh.jpg")
plt.show()
```

## Scatter Plots

```python
plt.scatter(num,squares,color="red",alpha=1.0,s=200)
plt.show()
```

## Histograms

```python
plt.hist(squares)
plt.show()
```

```python
x=[2,2,3,3,3,5,5,5,5,5,6,6,6,7,8,8,9,9,9,9,9,9]
plt.hist(x)
plt.show()
```

```python
plt.hist(x,bins=3)
plt.show()
```

```python
plt.hist(x,alpha=0.4)
plt.show()
```

## Pie Chart

```python
plt.pie(x,labels=x)
plt.show()
```

```python
plt.pie(x,labels=x,colors=("red","blue"))
plt.show()
```

## Box Plot

```python
plt.boxplot(squares)
plt.show()
```
