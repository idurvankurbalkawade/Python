## Basic Map

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
f1=plt.axes(projection=ccrs.PlateCarree())
f1.coastlines()
plt.show()
```

```python
fig=plt.figure()
f2=fig.add_subplot(projection=ccrs.PlateCarree())
f2.coastlines()
```

## Change Size

```python
fig=plt.figure(figsize=(10,10))
f2=fig.add_subplot(projection=ccrs.PlateCarree())
f2.coastlines()
```
