```
def pattern(n):
  num=1
  for i in range(0,n):
    for j in range(0,i+1):
      print(num,end=" ")
    num=num+1
    print("\r")

n=5
pattern(n)
```

```
def contnum(n):
    num = 1
    for i in range(0, n):
     
        for j in range(0, i+1):
          print(num, end=" ")
          num = num + 1
     
        
        print("\r")
 
n = 5
contnum(n)
```

```
num = int(input("Enter a number: "))
if num > 1:
   for i in range(2,num):
       if (num % i) == 0:
           print(num,"is not a prime number")
           break
   else:
       print(num,"is a prime number")
```
