## Scatter Plots

```python
import matplotlib.pyplot as plt
arr1=[1,2,3,4,5,6]
arr2=[55,65,80,90,140,210]
plt.scatter(arr1,arr2)
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_csv('height_weight.csv')
fig,ax=plt.subplots(figsize=(12,7))
ax.scatter(x=df['height'],y=df['female'])
plt.xlabel("Height of female")
plt.ylabel("Weight for female")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt 
fig,ax=plt.subplots(figsize=(12,7))
ax.scatter(x=df['height'],y=df['male'])
plt.xlabel("Height of male")
plt.ylabel("Weight for male")
plt.show()
```

```python
import matplotlib.pyplot as plt 
import pandas as pd
d1=pd.read_csv('marchcorona.csv')
fig,ax=plt.subplots(figsize=(11,5))
ax.scatter(x=d1['Day'],y=d1['Cases'])
plt.xlabel("Days")
plt.ylabel("Corona Cases")
plt.show()
```

## Box Plot

```python
import numpy as np
sales=np.array([6,7,13,17,20,25,39,41,43,49,51,62])
fig=plt.figure(figsize=(10,6))
plt.boxplot(sales)
plt.show()
```
