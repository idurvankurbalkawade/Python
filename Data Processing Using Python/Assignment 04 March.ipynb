```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Q1) 1st code
y=np.array([50,50])
subject=["Girls","Boys"]
plt.pie(y,labels=subject)
plt.legend()
plt.show()
```

```python
#Q2 2nd Code
arr=np.array([20,40,40])
trans=["Cycle","Walk","Bus or Car"]
plt.pie(arr,labels=trans)
plt.legend()
plt.show()
```
