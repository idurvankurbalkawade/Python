```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.array([50,25,15,10])
prg=np.array(["Entertainment","Sports","News","Informative"])
myexplode=[0,0,0.2,0]
plt.pie(x,labels=prg,explode=myexplode)
plt.legend(title="Info%")
plt.show()
```

```python
x=np.array([50,25,15,10])
prg=np.array(["Entertainment","Sports","News","Informative"])
mycolors=["red","blue","green","yellow"]
myexplode=[0,0,0.2,0]
plt.pie(x,labels=prg,explode=myexplode,colors=mycolors)
plt.legend(title="Info%")
plt.show()
```
