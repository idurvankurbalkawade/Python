## Image processing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy import misc
panda=misc.face()
plt.imshow(panda)
plt.show()
```

```python
from scipy import misc
panda=misc.face()
flip=np.flipud(panda)
plt.imshow(flip)
plt.show()
```

```python
from scipy import ndimage,misc
panda=misc.face()
pd_rotate=ndimage.rotate(panda,130)
plt.imshow(pd_rotate)
plt.show()
```

```python
from PIL import Image
imgob=Image.open("ironman.jpg") #open an existing image
flip=imgob.transpose(Image.FLIP_LEFT_RIGHT) # flip of left and right
imgob.show()    #show original image
flip.show()     #show horizontal flipped image
```

```python
df=pd.read_csv("product.csv")
x=df["price"].mean()
df["price"].fillna(x,inplace=True)
print(df)
```

```python
df=pd.read_csv("product.csv")
x=df["price"].median()
df["price"].fillna(x,inplace=True)
print(df)
```

```python
dp=pd.read_csv("product.csv")
for i in dp.index:
    if dp.loc[i,"price"]>200:
        dp.loc[i,"price"]=400
print(dp)
```

```python
dp=pd.read_csv("product.csv")
for i in dp.index:
    if dp.loc[i,"price"]<20:
        dp.drop(i,inplace=True)
print(dp)
```
