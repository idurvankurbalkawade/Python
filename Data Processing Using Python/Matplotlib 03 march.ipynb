```python
#no need to import libraries in each code you can import them once and can use in any code block
import matplotlib.pyplot as plt
import numpy as np
```

```python
#sample example
xpoints=np.array([0,10])
ypoints=np.array([0,50])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#plotting of points
xpoints=np.array([2,10])
ypoints=np.array([5,50])
plt.plot(xpoints,ypoints,"o")
plt.show()
```

```python
##Creating Bar Chart
a1=np.array(["DBMS","C++","OS","Python"])
a2=np.array([87,58,76,71])
plt.bar(a1,a2)
plt.show()
```

```python
##Creating Bar Chart
a1=np.array(["DBMS","C++","OS","Python"])
a2=np.array([87,58,76,71])
plt.bar(a1,a2,color="red")
plt.show()
```

```python
#adjusting width
##Creating Bar Chart
a1=np.array(["DBMS","C++","OS","Python"])
a2=np.array([87,58,76,71])
plt.bar(a1,a2,width=0.2,color="green")
plt.show()
```

```python
y=np.array([40,20,25,15])
subject=["C","C++","Python","Java"]
plt.pie(y,labels=subject)
plt.legend(title="Subject %")
plt.show()
```

```python
#horizontal bar
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```
